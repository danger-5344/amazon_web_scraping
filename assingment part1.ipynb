```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import visibility_of_element_located
```

```python
def web_page(url):
    browser = webdriver.Chrome()  # start a web browser
    browser.get(url)  # navigate to URL
    # wait for page to load
    # by waiting for <h1> element to appear on the page
    title = (
        WebDriverWait(driver=browser, timeout=10)
        .until(visibility_of_element_located((By.CSS_SELECTOR, "body")))
        .text
    )
    # retrieve fully rendered HTML content
    webpage = browser.page_source
    browser.close()
    return webpage

def product_name(soup):
    try:
        p_name=soup.find("span",attrs={'id':'productTitle'})
        p_name_value=p_name.text
        p_name_string=p_name_value.strip()
    except AttributeError:
        p_name_string=""
    return p_name_string
def product_price(soup):
    try:
        p_price=soup.find("span",attrs={'class':'a-price-whole'})
        p_price_value=p_price.text
        p_price_string=p_price_value.strip()
    except AttributeError:
        p_price_string=""
    return p_price_string
    
def product_rate(soup):
    try:
        p_rate=soup.find("i",attrs={'class':'a-icon a-icon-star a-star-4 cm-cr-review-stars-spacing-big'}).string.strip()
    except AttributeError:
        
        try:
            p_rate = soup.find("span", attrs={'class':'a-icon-alt'}).string.strip()
        except:
            p_rate = ""
    return p_rate

def product_review(soup):
    try:
        p_review=soup.find("span",attrs={'id':'acrCustomerReviewText'})
        p_review_value=p_review.text
        p_review_string=p_review_value.strip()
    except AttributeError:
        p_review_string=""
    return p_review_string
    
    
```

```python

d = {"PRODUCT_URL":[], "PRODUCT_NAME":[], "PRODUCT_PRICE":[], "RATING":[],"NO_OF_REVIEWS":[]}
for i in range(1,21):
    url="https://www.amazon.in/s?k=bags&page="+str(i)+"&crid=2M096C61O4MLT&qid=1692657449&sprefix=ba%2Caps%2C283&ref=sr_pg_1"
    webpage = web_page(url)
    soup = BeautifulSoup(webpage, "html.parser")
    links = soup.find_all("a", attrs={'class':'a-link-normal s-no-outline'})
    links_list=[]
    for link in links:
        links_list.append(link.get('href'))
    print(links_list)
    for link in links_list:
        if link[0:5]=="https":
            url=link.strip()
        else:
            url="https://www.amazon.in"+link
        browser = webdriver.Chrome()  # start a web browser
        browser.get(url)  # navigate to URL
        # wait for page to load
        # by waiting for <h1> element to appear on the page
        title = (WebDriverWait(driver=browser, timeout=10).until(visibility_of_element_located((By.CSS_SELECTOR, "h1"))).text)
        # retrieve fully rendered HTML content
        content = browser.page_source
        browser.close()
    
        soup = BeautifulSoup(content, "html.parser")
        d['PRODUCT_URL'].append("https://www.amazon.in"+link)
        d['PRODUCT_NAME'].append(product_name(soup))
        d['PRODUCT_PRICE'].append(product_price(soup))
        d['RATING'].append(product_rate(soup))
        d['NO_OF_REVIEWS'].append(product_review(soup))
    
amazon_df = pd.DataFrame.from_dict(d)
amazon_df['PRODUCT_NAME'].replace('', np.nan, inplace=True)
amazon_df = amazon_df.dropna(subset=['PRODUCT_NAME'])
amazon_df.to_csv("amazon_data.csv", header=True, index=False)
print('completed')
    

    
        
        
        

    

    
```
